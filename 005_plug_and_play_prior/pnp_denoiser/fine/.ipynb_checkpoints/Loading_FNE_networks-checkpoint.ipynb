# FNE networks

In this notebook, we provide elementary code to load / test approximately firmly nonexpansive (FNE) networks. We trained the model with different values of $\lambda$ (which we ca) in (3.10) of [Learning Maximally Monotone Operators for Image Recovery](https://arxiv.org/pdf/2012.13247.pdf). We load each network and test the reconstruction result.

The Lipschitz constant is computed with a power method. The FNE constant is the value of $\|\nabla J - \operatorname{Id}\|$, where $J$ is the considered DNN.

Training noise (n_lev) | $\lambda$ (ljr) | Lip. constant | FNE constant 
--- | --- | --- | --- 
2.0 | 0.0 | 5.33 | 13.30 
2.0 | 0.001 | 0.97 | 1.15 
2.0 | 0.002 | 0.96 | 1.04 
2.25 | 0.0 | 6.07 | 23.82 
2.25 | 0.001 | 0.97 | 1.32 
2.25 | 0.002 | 0.96 | 1.01 
2.5 | 0.0 | 5.03 | 12.94 
2.5 | 0.001* | 0.97 | 2.92 
2.5 | 0.005| 0.96 | 1.04 

*This value of $\lambda$ appears suboptimal, I will retrain for a higher value.

## 1. Loading the network

We first show how to load our Pytorch DnCNN.

```python
import torch
import torch.nn as nn

name_file = 'DNCNN_nch_1_sigma_2.25_ljr_0.001'
model_weights = torch.load('ckpts/finetuned/'+name_file+'.ckpt', map_location=torch.device('cpu'))
```

```python
class DnCNN(nn.Module):
    def __init__(self, nc_in, nc_out, depth, act_mode, bias=True, nf=64):
        super(DnCNN, self).__init__()

        self.depth = depth

        self.in_conv = nn.Conv2d(nc_in, nf, kernel_size=3, stride=1, padding=1, bias=bias)
        self.conv_list = nn.ModuleList(
            [nn.Conv2d(nf, nf, kernel_size=3, stride=1, padding=1, bias=bias) for _ in range(self.depth - 2)])
        self.out_conv = nn.Conv2d(nf, nc_out, kernel_size=3, stride=1, padding=1, bias=bias)

        if act_mode == 'R':  # Kai Zhang's nomenclature
            self.nl_list = nn.ModuleList([nn.ReLU() for _ in range(self.depth - 1)])

    def forward(self, x_in):

        x = self.in_conv(x_in)
        x = self.nl_list[0](x)

        for i in range(self.depth - 2):
            x_l = self.conv_list[i](x)
            x = self.nl_list[i + 1](x_l)

        return self.out_conv(x) + x_in
```

```python
cuda = True if torch.cuda.is_available() else False
Tensor = torch.cuda.FloatTensor if cuda else torch.FloatTensor

avg, bn, depth = False, False, 20
net = DnCNN(1, 1, depth, 'R')

if cuda:
    model = nn.DataParallel(net).cuda()
else:
    model = nn.DataParallel(net)
    
model.module.load_state_dict(model_weights["state_dict"], strict=True)

model.eval() 
```

We wrap that in a function

```python
def load_dncnn_weights(n_ch, n_lev, ljr):

    name_file = 'DNCNN_nch_'+str(n_ch)+'_sigma_'+str(n_lev)+'_ljr_'+str(ljr)
    
    model_weights = torch.load('ckpts/finetuned/'+name_file+'.ckpt', map_location=torch.device('cpu'))

    cuda = True if torch.cuda.is_available() else False
    Tensor = torch.cuda.FloatTensor if cuda else torch.FloatTensor

    avg, bn, depth = False, False, 20
    net = DnCNN(1, 1, depth, 'R')

    if cuda:
        model = nn.DataParallel(net).cuda()
    else:
        model = nn.DataParallel(net)

    model.module.load_state_dict(model_weights["state_dict"], strict=True)
    model.eval() 
    
    return model
```

## 2. Testing a DnCNN on an image

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio

cuda = True if torch.cuda.is_available() else False
Tensor = torch.cuda.FloatTensor if cuda else torch.FloatTensor

def get_image(path='test_pictures/koala.jpg', even=False, n_channels=3, n_max=None):
    '''
    Gets a single image and casts it to torch
    '''
    im_np = imageio.imread(path)/255.
    im_np = np.moveaxis(im_np,-1,0)
    sample = torch.from_numpy(im_np).type(Tensor)

    sample.unsqueeze_(0)

    x_bd = 2*(sample.shape[-2]//2) if even else -1
    y_bd = 2*(sample.shape[-1]//2) if even else -1
    
    if n_channels==1:
        sample = sample.mean(axis=1)
        sample.unsqueeze_(0)

    return sample[...,:x_bd, :y_bd]

def torch_psnr(est, true, max_peak = 1.):
    mse = torch.mean((est - true) ** 2) # Taken from https://github.com/huster-wgm/Pytorch-metrics/blob/master/metrics.py
    return 10 * torch.log10(max_peak/ mse)

im = get_image(path='test_pictures/koala.jpg', n_channels=1)

im_np = im.detach().numpy()
print('Loaded image of shape: ', im_np.shape)
```

```python
im_noisy = im + 2.25/255.*torch.randn_like(im)

out = model(im_noisy)
```

```python
im_noisy_np = im_noisy.detach().numpy()[0,0]
out_np = out.detach().numpy()[0,0]
im_np = im.detach().numpy()[0,0]

plt.figure(figsize=(15,15))
plt.subplot(131)
plt.imshow(im_noisy_np[:32,:32])
plt.title('Input noisy image')
plt.subplot(132)
plt.imshow(out_np[:32,:32])
plt.title('Output image')
plt.subplot(133)
plt.imshow(im_np[:32,:32])
plt.title('Groundtruth')
plt.show()
```

## 3. Comparing DnCNNs trained with different FNE regularisations

We trained the model with different values of $\lambda$ in (3.10) of [Learning Maximally Monotone Operators for Image Recovery](https://arxiv.org/pdf/2012.13247.pdf). We load each network and test the reconstruction result.

### 3.1 Noise level 2.0/255

```python
n_ch, n_lev = 1, 2.0

torch.manual_seed(0)  # Reproducibility
im_noisy = im + n_lev/255.*torch.randn_like(im)

im_noisy_np = im_noisy.detach().numpy()[0,0]
im_np = im.detach().numpy()[0,0]

torch_psnr_noisy = torch_psnr(im_noisy, im_np)

plt.figure(figsize=(15,10))
plt.subplot(231)
plt.imshow(im_noisy_np[:32,:32])
plt.title('Input noisy image, psnr = {:.2f}'.format(torch_psnr_noisy))
plt.subplot(232)
plt.imshow(im_np[:32,:32])
plt.title('Groundtruth')

for idx, ljr in enumerate([0.0, 0.001, 0.002]):
    
    model = load_dncnn_weights(n_ch, n_lev, ljr)
    out = model(im_noisy).detach()
    
    out_np = out.detach().numpy()[0,0]
    
    torch_psnr_out = torch_psnr(out, im_np)
    
    plt.subplot(int('23'+str(3+idx)))
    plt.imshow(out_np[:32,:32])
    plt.title('Output, $\lambda$ = '+str(ljr)+', psnr = {:.2f}'.format(torch_psnr_out))
    
plt.show()
```
